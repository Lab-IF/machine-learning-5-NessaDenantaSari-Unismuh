```python
# ============================================================
# IMPORT LIBRARY
# ============================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

plt.style.use("seaborn-v0_8")
```

```python
# ============================================================
# TUGAS 1: ARRAY RANDOM 20 ANGKA + STATISTIK
# ============================================================
print("\n=== Tugas 1: Statistik Array Random ===")

np.random.seed(42)
arr = np.random.rand(20) * 100

print("Array :", np.round(arr, 2))
print("Mean  :", round(arr.mean(), 2))
print("Median:", round(np.median(arr), 2))
print("Std   :", round(arr.std(), 2))
print("Min   :", round(arr.min(), 2))
print("Max   :", round(arr.max(), 2))
print("Range :", round(arr.max() - arr.min(), 2))

# Analisis tambahan distribusi
above_mean = np.sum(arr > arr.mean())
below_mean = np.sum(arr <= arr.mean())

print("\nDistribusi terhadap Mean:")
print("Jumlah di atas mean :", above_mean)
print("Jumlah di bawah mean:", below_mean)
```

```python
# ============================================================
# TUGAS 2: DATAFRAME 10 MAHASISWA + TOP 3
# ============================================================
print("\n=== Tugas 2: DataFrame & Top 3 Nilai ===")

np.random.seed(0)

df_mhs = pd.DataFrame({
    "Nama": [f"Mahasiswa_{i+1}" for i in range(10)],
    "Nilai": np.random.randint(50, 100, 10)
})

# Tambah kategori kelulusan
df_mhs["Kategori"] = np.where(df_mhs["Nilai"] >= 75, "Lulus", "Remedial")

print("\nData Mahasiswa:")
print(df_mhs)

print("\nStatistik Nilai:")
print(df_mhs["Nilai"].describe())

top3 = df_mhs.nlargest(3, "Nilai")

print("\nTop 3 Nilai Tertinggi:")
print(top3)

print("\nPersentase Kelulusan:")
print(df_mhs["Kategori"].value_counts(normalize=True) * 100)
```

```python
# ============================================================
# TUGAS 3: VISUALISASI DATA NILAI
# ============================================================
print("\n=== Tugas 3: Visualisasi ===")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Grafik 1: Bar Chart + Garis Mean
axes[0].bar(df_mhs["Nama"], df_mhs["Nilai"], color="steelblue")
axes[0].axhline(df_mhs["Nilai"].mean(), color="red", linestyle="--", label="Mean")
axes[0].set_title("Bar Chart Nilai Mahasiswa")
axes[0].set_xlabel("Nama")
axes[0].set_ylabel("Nilai")
axes[0].tick_params(axis="x", rotation=45)
axes[0].legend()

# Grafik 2: Histogram Distribusi
axes[1].hist(df_mhs["Nilai"], bins=5, color="coral", edgecolor="black")
axes[1].set_title("Histogram Distribusi Nilai")
axes[1].set_xlabel("Nilai")
axes[1].set_ylabel("Frekuensi")

plt.tight_layout()
plt.savefig("pertemuan01.png")
plt.close()

print("Grafik disimpan ke pertemuan01.png")
```

```python
# ============================================================
# TUGAS 4: EKSPLORASI DATASET WINE
# ============================================================
print("\n=== Tugas 4: Dataset Wine ===")

wine = load_wine(as_frame=True)
df_wine = wine.frame

print("Ukuran data :", df_wine.shape)
print("Jumlah fitur:", len(wine.feature_names))
print("Jumlah kelas:", df_wine["target"].nunique())
print("Kolom:", list(df_wine.columns))

# Statistik 3 fitur pertama
print("\nStatistik 3 Fitur Pertama:")
print(df_wine[wine.feature_names[:3]].agg(["mean", "min", "max"]))
```

```python
# ============================================================
# INSIGHT SINGKAT
# ============================================================
print("\nInsight 1: Dataset memiliki",
      df_wine.shape[0], "sampel dan",
      df_wine.shape[1] - 1, "fitur numerik.")

print("Insight 2: Target terdiri dari",
      df_wine['target'].nunique(),
      "kelas berbeda (klasifikasi multi-class).")

print("Insight 3: Rata-rata kandungan alcohol =",
      round(df_wine['alcohol'].mean(), 2),
      "yang menunjukkan variasi karakteristik kimia antar sampel.")

print("\n✅ Pertemuan 01 selesai.")
```
